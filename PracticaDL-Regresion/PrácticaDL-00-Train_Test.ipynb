
#**División en train/test.**

Dividimos el Dataset de airbnb de Madrid en dos partes, entrenamiento y test, para medir la capacidad de generalización. 
Utilizamos un 70-75% de las muestras para entrenamiento y 30-25% para el test.


```
# Primero cargamos librerías y funciones necesarias
import numpy as np 
import pandas as pd

#Para trabajar en el colab
from google.colab import drive
drive.mount('/content/drive')

import matplotlib.pyplot as plt # para dibujar
%matplotlib inline

from sklearn.model_selection import train_test_split

# Carga del dataset de pisos airbnb Madrid
# El fichero lo ubico directamente en el directorio de trabajo 
airbnbMadrid = pd.read_csv('/content/drive/My Drive/BootCamp - BigDataIV - DL/PracticaDL-Regresion/airbnb-listings Madrid-v1.csv', sep=";", decimal='.')

# Dividimos el dataset cargado en train y test
train, test = train_test_split(airbnbMadrid, test_size=0.2, shuffle=True, random_state=0)
# Muestra por pantalla en número de registros y columnas de cada dataset
print(f'Dimensiones del dataset de training : {train.shape}') 
print(f'Dimensiones del dataset de test     : {test.shape}')  

# Guardamos en sendos archivos de salida en el mismo directorio
train.to_csv('/content/drive/My Drive/BootCamp - BigDataIV - DL/PracticaDL-Regresion/train.csv', sep=';', decimal='.', index=False) # registros 10616
test.to_csv('/content/drive/My Drive/BootCamp - BigDataIV - DL/PracticaDL-Regresion/test.csv', sep=';', decimal='.', index=False) # registros 2655
```
