```
# Sustituimos los missings por la media en caso que la variable(numérica) tenga nulos : 
airbnbMadrid_train["Bedrooms"] = airbnbMadrid_train["Bedrooms"].fillna(airbnbMadrid_train["Bedrooms"].mean())
airbnbMadrid_train["Bathrooms"] = airbnbMadrid_train["Bathrooms"].fillna(airbnbMadrid_train["Bathrooms"].mean())
airbnbMadrid_train["Beds"] = airbnbMadrid_train["Beds"].fillna(airbnbMadrid_train["Beds"].mean())
airbnbMadrid_train["Square Feet"] = airbnbMadrid_train["Square Feet"].fillna(airbnbMadrid_train["Square Feet"].mean())
airbnbMadrid_train["Price"] = airbnbMadrid_train["Price"].fillna(airbnbMadrid_train["Price"].mean())
airbnbMadrid_train["Weekly Price"] = airbnbMadrid_train["Weekly Price"].fillna(airbnbMadrid_train["Weekly Price"].mean())
airbnbMadrid_train["Monthly Price"] = airbnbMadrid_train["Monthly Price"].fillna(airbnbMadrid_train["Monthly Price"].mean())
airbnbMadrid_train["Security Deposit"] = airbnbMadrid_train["Security Deposit"].fillna(airbnbMadrid_train["Security Deposit"].mean())
airbnbMadrid_train["Cleaning Fee"] = airbnbMadrid_train["Cleaning Fee"].fillna(airbnbMadrid_train["Cleaning Fee"].mean())
airbnbMadrid_train["Review Scores Rating"] = airbnbMadrid_train["Review Scores Rating"].fillna(airbnbMadrid_train["Review Scores Rating"].mean())
airbnbMadrid_train["Review Scores Accuracy"] = airbnbMadrid_train["Review Scores Accuracy"].fillna(airbnbMadrid_train["Review Scores Accuracy"].mean())
airbnbMadrid_train["Review Scores Cleanliness"] = airbnbMadrid_train["Review Scores Cleanliness"].fillna(airbnbMadrid_train["Review Scores Cleanliness"].mean())
airbnbMadrid_train["Review Scores Checkin"] = airbnbMadrid_train["Review Scores Checkin"].fillna(airbnbMadrid_train["Review Scores Checkin"].mean())
airbnbMadrid_train["Review Scores Communication"] = airbnbMadrid_train["Review Scores Communication"].fillna(airbnbMadrid_train["Review Scores Communication"].mean())
airbnbMadrid_train["Review Scores Value"] = airbnbMadrid_train["Review Scores Value"].fillna(airbnbMadrid_train["Review Scores Value"].mean())
airbnbMadrid_train["Review Scores Location"] = airbnbMadrid_train["Review Scores Location"].fillna(airbnbMadrid_train["Review Scores Location"].mean())
airbnbMadrid_train["Calculated host listings count"] = airbnbMadrid_train["Calculated host listings count"].fillna(airbnbMadrid_train["Calculated host listings count"].mean())
airbnbMadrid_train["Reviews per Month"] = airbnbMadrid_train["Reviews per Month"].fillna(airbnbMadrid_train["Reviews per Month"].mean())
```

```
# Convertimos la variable de superficie de pies al cuadrado a metros al cuadrado.
feetFeatures = ['Square Feet']
airbnbMadrid_train[feetFeatures] = airbnbMadrid_train[feetFeatures].apply(lambda x: x * 0.3048 * 0.3048)
#Renombramos la columna
airbnbMadrid_train = airbnbMadrid_train.rename(columns={'Square Feet':'SquareM2'})
```

## Categorización de varibles.
Usamos el método mean encoder para las variables no categorizadas.


```
#Hay 15 variables a categorizar : float64(18), int64(10), object(15)

mean_encode = airbnbMadrid_train.groupby('Neighbourhood')['Price'].mean()
airbnbMadrid_train.loc[:,'Neighbourhood_mean_enc']=airbnbMadrid_train['Neighbourhood'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Neighbourhood Cleansed')['Price'].mean()
airbnbMadrid_train.loc[:,'Neighbourhood_Cleansed_mean_enc']=airbnbMadrid_train['Neighbourhood Cleansed'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Neighbourhood Group Cleansed')['Price'].mean()
airbnbMadrid_train.loc[:,'Neighbourhood_Group_Cleansed_mean_enc']=airbnbMadrid_train['Neighbourhood Group Cleansed'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('State')['Price'].mean()
airbnbMadrid_train.loc[:,'State_mean_enc']=airbnbMadrid_train['State'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Zipcode')['Price'].mean()
airbnbMadrid_train.loc[:,'Zipcode_mean_enc']=airbnbMadrid_train['Zipcode'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Market')['Price'].mean()
airbnbMadrid_train.loc[:,'Market_mean_enc']=airbnbMadrid_train['Market'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Smart Location')['Price'].mean()
airbnbMadrid_train.loc[:,'Smart_Location_mean_enc']=airbnbMadrid_train['Smart Location'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Country Code')['Price'].mean()
airbnbMadrid_train.loc[:,'Country_Code_mean_enc']=airbnbMadrid_train['Country Code'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Country')['Price'].mean()
airbnbMadrid_train.loc[:,'Country_mean_enc']=airbnbMadrid_train['Country'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Property Type')['Price'].mean()
airbnbMadrid_train.loc[:,'Property_Type_mean_enc']=airbnbMadrid_train['Property Type'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Room Type')['Price'].mean()
airbnbMadrid_train.loc[:,'Room_Type_mean_enc']=airbnbMadrid_train['Room Type'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Bed Type')['Price'].mean()
airbnbMadrid_train.loc[:,'Bed_Type_mean_enc']=airbnbMadrid_train['Bed Type'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Calendar Updated')['Price'].mean()
airbnbMadrid_train.loc[:,'Calendar_Updated_mean_enc']=airbnbMadrid_train['Calendar Updated'].map(mean_encode)

mean_encode = airbnbMadrid_train.groupby('Cancellation Policy')['Price'].mean()
airbnbMadrid_train.loc[:,'Cancellation_Policy_mean_enc']=airbnbMadrid_train['Cancellation Policy'].map(mean_encode)
```
